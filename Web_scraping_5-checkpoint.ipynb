a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
FTeam_name=soup.find('tr',class_="rankings-block__banner")
FTeam_name.get_text(strip=True,separator=" ").split(" ")
```

```python
otherTeam_name=soup.find_all('tr',class_="table-body")
otherTeam_name
```

```python
final_Team_lst=[]
final_Team_lst.append(FTeam_name.get_text(strip=True,separator=" ").split(" "))
final_Team_lst
```

```python
for i in otherTeam_name:
    split_lst=i.get_text(strip=True,separator=" ").split(" ")
    final_Team_lst.append(split_lst)
final_Team_lst[:10]
```

b) Top 10 ODI Batsmen along with the records of their team and rating.

```python
page=requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
fdata=soup.find("div",class_="rankings-block__top-player")
fdata.get_text(strip=True,separator=" ").split(" ")
```

```python
other_data=soup.find_all("tr",class_="table-body")
other_data
```

```python
final_lst=[]
final_lst.append(fdata.get_text(strip=True,separator=" ").replace("This player has moved up in the rankings since the previous rankings update ",""))
final_lst
```

```python
for i in other_data:
    split_lst=i.get_text(strip=True,separator=" ").replace("This player has moved up in the rankings since the previous rankings update ","")
    split_lst=split_lst.replace("This player has moved down in the rankings since the previous rankings update ","")
    final_lst.append(split_lst)
final_lst[:10]
```

```python
first_data_bowl=soup.find_all("div",class_="rankings-block__top-player")

final_lst_bowl=[]

k=0
for i in first_data_bowl:
    k=k+1
    if k==2:
        split_lst_bowl=i.get_text(strip=True,separator=" ").split(" ")
        final_lst_bowl.append(split_lst_bowl)
final_lst_bowl
```

```python
final_lst_bowl=final_lst_bowl+final_lst[10:19]
final_lst_bowl
```
